```python
import ipywidgets as widgets
from IPython.display import display,HTML

# 1. Create an Output widget to capture prints
out = widgets.Output()

def callback(b):
    # 2. Use the output widget context
    with out:
        #out.clear_output() # Optional: clears previous message
        new_svg = f'<p>Iterations: {20}</p>'
        new_svg += f'<p style="color:green">Green indicates no occupants</p>'
        out.display(HTML("<div id='sim-container'>Loading...</div>"), display_id=True)
        out.update(HTML(f"<div id='sim-container'>{new_svg}</div>"))
        

start_button = widgets.Button(
    description='▶ Start Calculation', 
    button_style='success'
)

start_button.on_click(callback)

# 3. Display both the button and the output area
display(start_button, out)
```
