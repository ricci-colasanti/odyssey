````python
from IPython.display import display, HTML,Javascript,clear_output
import random as rnd
import time
import json
from svg_canvas import SVGCanvas
````

### The IPython.display.display() function allows you to:

Force output anywhere in your code (not just the last line).  
Choose the format (HTML, Markdown, Image, Audio, Latex, etc.).  
Control the behavior (append, replace, or update).  


By default, every time you call display(), Jupyter creates a new, unique output block in the cell. If you call it 10 times in a loop, you get 10 blocks stacked vertically.   

When you set display_id=True, you are telling Jupyter:  

*"Don't create a new block. Instead, assign a unique ID to this output. If I call display() again with the same ID later, update this specific block instead of adding a new one."*   
 ``` python
 output_display.update(HTML(f"<div id='sim-container'>{new_svg}</div>"))
 ```



````python
# 1. Initialize
svg = SVGCanvas(500, 500, 60)
vertices = [
    [25, 12], [32, 10], [38, 15], [35, 22], [28, 25], [22, 20]
]

# 2. Create a placeholder output
output_display = display(HTML("<div id='sim-container'>Loading...</div>"), display_id=True)

# 3. The Loop
for i in range(40):
    # --- Your Logic ---
    svg.clear()
    svg.addPolygon(vertices,'rgb(200,255,200)')
    svg.addCircle(5.440926694200909, 6.254578264027191,'rgb(255, 200, 200)')
    svg.addRect(5,5,'rgb(255, 100, 100)')
    svg.addRect(i,i,'blue')
    
    new_svg = svg.getCanvas()
    
    # --- Update the EXISTING output ---
    # display_id=True allows us to update this specific output in place
    # No clear_output needed! No blank lines! No flicker!
    output_display.update(HTML(f"<div id='sim-container'>{new_svg}</div>"))
    
    time.sleep(0.1)
````
